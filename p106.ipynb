```python
import csv
import plotly.express as px
import numpy as np

with open('correlation3.csv') as csv_files:
  df=csv.DictReader(csv_files)
  fig = px.scatter(df, x='Days Present', y='Marks In Percentage')
  fig.show()

def getDataSource(data_path):
  days = []
  marks = []
  with open(data_path) as csv_files:
    csv_reader = csv.DictReader(csv_files)
    for row in csv_reader:
      days.append(float(row['Days Present']))
      marks.append(float(row['Marks In Percentage']))
  
  return{
      'x': days,
      'y': marks
  }

def findCorrelation(dataSource):
  relation = np.corrcoef(dataSource['x'], dataSource['y'])
  print('correlation between days present and marks is', relation[0,1])

data_path = 'correlation3.csv'
dataSource = getDataSource(data_path)
findCorrelation(dataSource)
```
